```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import ndimage
from scipy import spatial

from skimage import filter as skfilter
from skimage import data, color, exposure
from skimage import measure
from skimage.filter import rank
from skimage.io import imread
from skimage.transform import hough_circle, hough_ellipse
from skimage.feature import peak_local_max
from skimage.draw import circle_perimeter, circle
from skimage import morphology
from skimage import segmentation
from skimage.morphology import disk, square
from skimage import draw
from skimage.feature import peak_local_max
from sklearn import svm
```

```python
# Read image, initial preprocessing.
orig_image = imread('/Users/flamholz/Documents/Rotations/Fletcher/example_images/several_GUVs_multiple_2way_interfaces.tif')
image = orig_image

glob = exposure.equalize_hist(image) * 255
glob = exposure.rescale_intensity(image)
loc = rank.enhance_contrast(image, disk(3))


# Subset with only the vesicles we care about.
image = image[150:280, 250:350]
glob = glob[150:280, 250:350]
loc = loc[150:280, 250:350]

plt.figure(figsize=(10, 10))

plt.subplot(131)
plt.imshow(image, cmap=plt.cm.gray, interpolation='nearest')
plt.axis('off')
plt.title('Image')

plt.subplot(132)
plt.imshow(glob, cmap=plt.cm.gray, interpolation='nearest')
plt.axis('off')
plt.title('Global Contrast Boost')

plt.subplot(133)
plt.imshow(loc, cmap=plt.cm.gray, interpolation='nearest')
plt.axis('off')
plt.title('Local Contrast Boost')
plt.show()
```

```python
image = glob
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10, 10))
ax.imshow(image, cmap=plt.cm.gray)

# Edge detect, detect radii
edges = skfilter.canny(image, sigma=3, low_threshold=10, high_threshold=50)
hough_radii = np.arange(15, 50, 2)
hough_res = hough_circle(edges, hough_radii)

# For each radius, extract two circles
centers = []
accums = []
radii = []

for radius, h in zip(hough_radii, hough_res):
    peaks = peak_local_max(h, num_peaks=2)
    centers.extend(peaks)
    accums.extend(h[peaks[:, 0], peaks[:, 1]])
    radii.extend([radius, radius])

# Draw the most prominent 5 circles
shape = image.shape
hough_mask = np.zeros((shape[0], shape[1], 3))
center_mask = np.zeros(shape)
for idx in np.argsort(accums)[::-1][:5]:
    center_x, center_y = centers[idx]
    radius = radii[idx]
    cx, cy = circle_perimeter(center_y, center_x, radius)
    in_frame_idx = np.where(np.logical_and(cx <= shape[0], cy <= shape[1]))
    center_mask[center_x, center_y] = 1

    hough_mask[cy[in_frame_idx], cx[in_frame_idx]] = (220, 20, 20)
    
ax.imshow(hough_mask, cmap=plt.cm.gray, alpha=0.3)
plt.show()
```

```python
# Now we want to separate the two objects in image
# Generate the markers as local maxima of the distance to the background
big_center_mask = skfilter.gaussian_filter(center_mask, 3) * 10
markers = ndimage.label(center_mask)[0]
watershed_labels = morphology.watershed(image, markers)

fig, axes = plt.subplots(ncols=2, figsize=(10, 20))
ax0, ax1 = axes

ax0.imshow(image, cmap=plt.cm.gray, interpolation='nearest')
ax0.imshow(big_center_mask, cmap=plt.cm.jet, alpha=0.3)
ax0.set_title('Overlapping objects')
ax1.imshow(image, cmap=plt.cm.gray, interpolation='nearest')
ax1.imshow(watershed_labels, cmap=plt.cm.spectral, interpolation='nearest', alpha=0.4)
ax1.set_title('Separated objects (Watershed)')

for ax in axes:
    ax.axis('off')

plt.subplots_adjust(hspace=0.01, wspace=0.01, top=1, bottom=0, left=0,
                    right=1)
plt.show()
```

```python
centr_getter = lambda x: x.centroid
area_getter = lambda x: x.area

props = measure.regionprops(watershed_labels)
props = filter(lambda x: x.area > 30, props) # No tiny regions
props = sorted(props, key=area_getter)       # Sort by area
props.pop(-1)                                # Assume biggest area is BG...

centroids = np.array(map(centr_getter, props))
centers = np.array(centers)
radii = np.array(radii)

distances = spatial.distance.cdist(centroids, centers)
closest_centers_idx = np.argmin(distances, axis=1)
closest_centers = centers[closest_centers_idx]
closest_radii = radii[closest_centers_idx]

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10, 10))
ax.imshow(image, cmap=plt.cm.gray)

perim_mask = np.zeros((shape[0], shape[1], 3))
fill_mask = np.zeros((shape[0], shape[1], 3))
for i in xrange(len(props)):
    c = closest_centers[i]
    r = closest_radii[i]
    cx, cy = circle(c[0], c[1], r-1)
    px, py = circle_perimeter(c[0], c[1], r)
    perim_mask[px, py] = (220, 20, 20)
    fill_mask[cx, cy] = (220, 220, 20)

ax.plot(closest_centers[:,1], closest_centers[:,0], 'g')    
ax.imshow(perim_mask, cmap=plt.cm.gray, alpha=0.3)
ax.imshow(fill_mask, cmap=plt.cm.gray, alpha=0.1)

plt.show()
```

```python
def boundary_angle_to_centroids(segment_data, centroids_vector):
    min_r, min_c, max_r, max_c = segment_data.bbox
    
    # Pad b/c we need a little background to find the segment boundary.
    padded_segment_image = np.pad(segment_data.filled_image, pad_width=4,
                                  mode='constant', constant_values=0)
    segment_boundary = segmentation.find_boundaries(padded_segment_image)
    segment_boundary = segment_boundary[4:-3,4:-3]  # unpad
    segment_centroid = np.array(segment_data.centroid)
        
    # Coordinates of the boundary of this segment
    boundary_idxs = np.where(segment_boundary)
    boundary_coords = np.array(boundary_idxs).T + [min_r, min_c]  # Coordinates in original image
    boundary_vectors = boundary_coords - segment_centroid         # Vectors to centroid
    
    # Calculate the absolute angle theta between boundary pixels
    # and the segment centroid. Theta is relative to the vector between 
    # the two vesicle centroids.
    dot_prod = boundary_vectors.dot(centroids_vector)
    centroid_v_norm = np.linalg.norm(centroids_vector)
    boundary_v_norms = np.linalg.norm(boundary_vectors, axis=1)
    normalized_dot = dot_prod / (centroid_v_norm * boundary_v_norms)
    theta_to_centroid = np.arccos(normalized_dot)
    theta_to_centroid = np.rad2deg(theta_to_centroid)
    
    # Determinant of 2x2 matrix of boundary and centroid vector determines
    # orientation. This allows us to give absolute angles to boundary points
    # relative to the centroid, i.e. ones that range from 0 to 360 degrees.
    bv_copy = boundary_vectors.copy()
    bv_copy[:,0] *= centroids_vector[1]
    bv_copy[:,1] *= centroids_vector[0]
    dets = bv_copy[:,0] - bv_copy[:,1]
    
    # Make the angles fully circular - (do not fully understand)
    absolute_thetas = theta_to_centroid.copy()
    absolute_thetas[np.where(dets > 0)] = 180.0 - theta_to_centroid[np.where(dets > 0)]
    absolute_thetas[np.where(dets > 0)] += 180.0   
    
    return absolute_thetas, boundary_idxs, boundary_coords

def get_intensities_along_boundary(image, absolute_thetas, boundary_coords):
    intensities = np.zeros(absolute_thetas.shape)
    for theta_idx in xrange(len(absolute_thetas)):
        coords = boundary_coords[theta_idx,:]
        intensities[theta_idx] = image[coords[0], coords[1]]
    return intensities
```

```python
# Vector between centroids of the two vesicles
centroids_vector = np.array(props[0].centroid) - props[1].centroid
mean_image = rank.mean(image, disk(4))

# Calculate intensities along the boundary as a function of angle.
thetas1, boundary_idxs1, boundary1 = boundary_angle_to_centroids(props[0], centroids_vector)
intensities1 = get_intensities_along_boundary(mean_image, thetas1, boundary1)
# Note: use -centroids_vector so that angles are oriented towards the centroid for both vesicles
thetas2, boundary_idxs2, boundary2 = boundary_angle_to_centroids(props[1], -centroids_vector)
intensities2 = get_intensities_along_boundary(mean_image, thetas2, boundary2)

segment_1_shape = props[0].image.shape
segment_2_shape = props[1].image.shape
thetas_in_image1 = np.zeros((segment_1_shape[0]+1, segment_1_shape[1]+1))
thetas_in_image2 = np.zeros((segment_2_shape[0]+1, segment_2_shape[1]+1))

thetas_in_image1[boundary_idxs1] = thetas1
thetas_in_image2[boundary_idxs2] = thetas2

plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(thetas_in_image1, cmap=plt.cm.jet)

plt.subplot(122)
plt.imshow(thetas_in_image2, cmap=plt.cm.jet)
plt.tight_layout()

plt.figure(figsize=(10, 10))
plt.title('Radial Intensity')
plt.plot(thetas1, intensities1, 'g.', label='Vesicle 1')
plt.plot(thetas2, intensities2, 'r.', label='Vesicle 2')
plt.plot([150, 150], [0, np.max(intensities1)], 'r--')
plt.plot([210, 210], [0, np.max(intensities1)], 'r--')
plt.legend()
plt.show()
```
