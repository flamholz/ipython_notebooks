```python
import pylab
import numpy as np
import csv
from scipy import integrate
```

```python
plate_rows = 'ABCDEFGH'
plate_cols = range(1,13)
all_cells = []
for row_label in plate_rows:
    for col_label in plate_cols:
        cell_label = row_label + str(col_label)
        all_cells.append(cell_label)
```

```python
cell_data = {}
timepoints = []
f = open('data/ben/2014106_ecoli_test_sgrnas 0-9 - OD Data Only.csv')
reader = csv.DictReader(f)

time_key = 'Time [s]'
for row in reader:
    timestamp_seconds = float(row[time_key])
    timepoints.append(timestamp_seconds)
    
    for cell_label in all_cells:
        cell_od = float(row[cell_label])
        cell_data.setdefault(cell_label, []).append(cell_od)

timepoints = np.array(timepoints)
timepoints_hours = timepoints / (60.0*60.0)
```

```python
neg_labels = 'E7,E8,E9,F10,F11,F12'.split(',')
pos_labels = 'B10,B11,B12,D10,D11,D12'.split(',')

pylab.figure()
pylab.title('Negatives (Shouldn\'t Grow Well)')
pylab.xlabel('time (hours)')
pylab.ylabel('OD')
for label in neg_labels:
    pylab.plot(timepoints / (60*60), cell_data[label])
    
    
pylab.figure()
pylab.title('Positives (Should Grow Well)')
pylab.xlabel('time (hours)')
pylab.ylabel('OD')
for label in pos_labels:
    pylab.plot(timepoints / (60*60), cell_data[label])
    
pylab.show()
```

```python
one_pos_cell = np.array(cell_data['B10'])
copy_cell = one_pos_cell.copy()
copy_cell.sort()
blank_od = np.mean(copy_cell[1:6])
one_pos_cell_blanked = one_pos_cell - blank_od

one_neg_cell = np.array(cell_data['E7'])
copy_cell = one_neg_cell.copy()
copy_cell.sort()
blank_od = np.mean(copy_cell[1:6])
one_neg_cell_blanked = one_neg_cell - blank_od
```

```python
last_index = pylab.where(timepoints_hours <= 24.0)[0][-1]
timepoints_to_24 = timepoints_hours[:last_index+1]
pos_od_to_24 = one_pos_cell_blanked[:last_index+1]
neg_od_to_24 = one_neg_cell_blanked[:last_index+1]

pylab.figure()
pylab.plot(timepoints_to_24, pos_od_to_24, 'b-')
pylab.plot(timepoints_to_24, neg_od_to_24, 'r-')
pylab.show()
```

```python
pos_auc = integrate.trapz(pos_od_to_24, timepoints_to_24)
neg_auc = integrate.trapz(neg_od_to_24, timepoints_to_24)
print 'Positive example AUC:', pos_auc
print 'Negative example AUC:', neg_auc
```
