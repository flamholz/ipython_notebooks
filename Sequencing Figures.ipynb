```python
import numpy as np
import pylab
```

```python
labels = ['A', 'B', 'C']
n_reads_total = 2 * np.array([2048120.0, 1041485.0, 1863466.0])
n_first_end_matching_insert = np.array([81866.0, 41564.0, 73051.0])
n_second_end_matching_insert = np.array([82023.0, 40528.0,  70489.0])
n_ends_matching_insert = n_first_end_matching_insert + n_second_end_matching_insert
n_first_end_matching_backbone = np.array([58085.0, 29509.0, 51680.0])
n_second_end_matching_backbone = np.array([58638.0, 28808.0, 50077.0])
n_ends_matching_backbone = n_first_end_matching_backbone + n_second_end_matching_backbone

pylab.figure(figsize=(20,10))
pylab.subplot(121)
pylab.bar(np.arange(len(labels)) + 0.25, n_reads_total / 1e6, width=0.5)
pylab.bar(np.arange(len(labels)) + 0.25, n_ends_matching_insert / 1e6, color='g', width=0.5)
pylab.bar(np.arange(len(labels)) + 0.25, n_ends_matching_backbone / 1e6, color='y', width=0.5)
pylab.xticks(np.arange(len(labels)) + 0.5, labels, fontsize=20)
pylab.ylabel('Millions of Paired Ends', fontsize=24)
pylab.xlabel('Sample', fontsize=24)

pylab.subplot(122)
pylab.bar(np.arange(len(labels)) + 0.25, 20.0*np.ones(len(labels)), width=0.5)
print n_ends_matching_insert / n_reads_total
print n_ends_matching_backbone / n_reads_total
pylab.bar(np.arange(len(labels)) + 0.25, 100.0*n_ends_matching_insert / n_reads_total, color='g',
          width=0.5)
pylab.bar(np.arange(len(labels)) + 0.25, 100.0*n_ends_matching_backbone / n_reads_total, color='y', width=0.5)
pylab.xticks(np.arange(len(labels)) + 0.5, labels, fontsize=20)
pylab.xlabel('Sample', fontsize=24)
pylab.ylabel('Percent of Paired Ends', fontsize=24)
pylab.legend()
pylab.show()
```
