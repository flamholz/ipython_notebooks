```python
import csv
import numpy as np
```

```python
spec_data_a = np.array([7.22, 3.32, 1.8, 0.88, 0.411, 0.21])
spec_data_b = np.array([6.88, 3.26, 1.74, 0.86, 0.423, 0.22])
dilutions = [2**i for i in range(6)]
expected_spec_data = np.array([7.0 / dil for dil in dilutions])
```

```python
from scipy.stats import linregress
f = open('/Users/flamholz/Downloads/Calibration - Sheet1.csv')
r = csv.reader(f)
r.next() # skip header

blanks = []
absorbances_per_dil = {}

for line in r:
    vals = [float(v) for v in line[1:]]
    tx = np.array(vals[0::2])
    rx = np.array(vals[1::2])
    absorbances = np.log10(rx/tx)
    line_desc = line[0].strip()
    if line_desc.startswith('blank'):
        blanks.append(absorbances)
    elif line_desc.endswith('x'):
        dilution = int(line_desc.strip('x'))
        absorbances_per_dil.setdefault(dilution, []).append(absorbances)
    else:
        assert False, line[0]

dilutions = sorted(absorbances_per_dil.keys())
mean_absorbances = []
std_absorbances = []
for dil in dilutions:
    absorbances = absorbances_per_dil[dil]
    mean_absorbances.append(np.mean(absorbances, axis=0))
    std_absorbances.append(np.std(absorbances, axis=0))

mean_blank = np.mean(blanks, axis=0)
mean_absorbances = np.array(mean_absorbances)
std_absorbances = np.array(std_absorbances)
blanked_absorbances = mean_blank - mean_absorbances
log_dilutions = np.log2(dilutions)
```
